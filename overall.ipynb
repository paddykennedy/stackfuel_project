```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
%matplotlib inline
import matplotlib.pyplot as plt
import sqlalchemy as sa
```

```python
df = pd.read_csv('tipsst10_tabular.tsv',sep='\t',na_values = ':')
df.head()
```

```python
import re
new_columns = df['freq,sectperf,unit,geo\\TIME_PERIOD'].str.split(',', expand=True)

new_columns.columns = ['freq', 'sectperf','unit', 'geo']
df = pd.concat([df, new_columns], axis=1)
print(df.head())

for i in range(1,28):
    df.iloc[:,i] = df.iloc[:,i].apply(lambda x: re.sub(r'[^0-9.]', '', x))
df

for i in range(1,28):
    df.iloc[:,i] = pd.to_numeric(df.iloc[:,i], errors='coerce')
df
```
